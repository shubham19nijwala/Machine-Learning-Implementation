```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
iris.sample(5)
```

### 
**Applying LABEL ENCODER**

```python
# Target Column
iris['species'].unique()
```

```python
# Target column is Categorical, converting it to numeric by proving labels
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y=le.fit_transform(iris['species'])
```

```python
le.classes_
```

```python
# Dependent feature
y
```

```python
# Independent features
X=iris.iloc[:,:-1]
```

```python
X,y
```

## 
**Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=19)
```

## Decision Tree Implementation

```python
from sklearn.tree import DecisionTreeClassifier
```

## PostPruning

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt,filled=True)
```

```python
dt_1=DecisionTreeClassifier(max_depth=2)
```

```python
dt_1.fit(X_train,y_train)
```

```python
tree.plot_tree(dt_1,filled=True)
```

##
**Prediction**

```python
y_pred=dt_1.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

## PrePruning

```python
parameter={
                       'criterion':['gini','entropy','log_loss'],
                        'splitter':['best','random'],
                        'max_depth':[2,3,4,5],
                        'max_features':['auto','sqrt','log2']
                     }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
dt=DecisionTreeClassifier()
gs=GridSearchCV(dt,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
gs.fit(X_train,y_train)
```

```python
gs.best_params_
```

##
**Prediction**

```python
y_pred1=gs.predict(X_test)
y_pred1
```

```python
accuracy_score(y_pred1,y_test)
```

```python
print(classification_report(y_pred1,y_test))
```
